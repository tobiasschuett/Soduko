```python
def fillSoduko(Problemmatrix):
    matrix = Problemmatrix
    Felder = []
    
    #-------------------Feldklasse erstellen ------------------------------------------
    
    class Feld:
        def __init__(self, box, werte, zeile, spalte):
            self.box = box
            self.werte = werte
            self.zeile = zeile
            self.spalte = spalte
            
    #---------------------------- Feld viertuell erstellen:------------------------------
    
    for i in range(9):              
        for j in range(9):
            
            if matrix[i][j] != 0:
                Feldwert = matrix[i][j]
            else:
                Feldwert = list(range(1,10))
            
            if i < 3:
                if j < 3:
                    boxwert = 1
                elif j < 6:
                    boxwert = 2
                else:
                    boxwert = 3
            elif i < 6: 
                if j < 3:
                    boxwert = 4
                elif j < 6:
                    boxwert = 5
                else:
                    boxwert = 6
            else:
                if j < 3:
                    boxwert = 7
                elif j < 6:
                    boxwert = 8
                else:
                    boxwert = 9
                    
            Felder.append(Feld(boxwert,Feldwert,i,j))
   
    #------------------- Lösungsalgorithmus Schritt 1 - offensichtliche Streichen:-------------------------
    
    def Schritt1(felder):
        for k in range (500):             
            p = 0                           #leere Felder
            for i in felder:
                if type(i.werte) != int:    #feld ist leer also kein int
                    if len(i.werte) != 1:   #noch mehr als eine zahl möglich für feld
                        p += 1              #+1 leeres feld
                        for x in felder:
                            if type(x.werte) == int: 
                                
                                if x.box == i.box:        
                                    if x.werte in i.werte:   #wert schon in box
                                        i.werte.remove(x.werte)
                                        
                                if x.zeile == i.zeile:       
                                    if x.werte in i.werte:     #wert schon in zeile
                                        i.werte.remove(x.werte)
                            
                                if x.spalte == i.spalte:     
                                    if x.werte in i.werte:      #wert schon in spalte
                                        i.werte.remove(x.werte)
                               
                    else:                             #eine mögliche zahl als int eintrag
                        i.werte = i.werte[0]
                        print("Feld ausgefüllt")
                    
            print("Schleifendurchlauf:",k,"leere Felder:",p)  # aktueller Durchlauf, Anzahl leere Felder
        
            if k > 0:            
                if p == leereFelder:         #es wurden keine felder gefüllt -> abbruch
                    print("1 Stufe fertig")
                    break
            leereFelder = p
        
            if p == 0:                       #keine leeren felder mehr -> abbruch
                print("komplett FERTIG")
                break
            
    Schritt1(Felder)        
                     
    #----------------Lösungsalgorithmus Schritt 2 - Feldoptionen vergleichen:---------------------------
    
    for n in range (100):
        FelderVoll = True
        
        for i in Felder: 
            if type(i.werte) != int:  #werte ist liste
                if len(i.werte) == 1: #liste nur ein wert
                    i.werte = i.werte[0]
                    print("Feld",Felder.index(i),"ausgefüllt mit Schritt 2-1, nochmal Schritt 1!")
                    Schritt1(Felder)              #NEUER EINTRAG, NOCHMAL NEU RAUSSTREICHEN -> SCHRITT 1
                else:                   #liste mehr als ein wert
                    FelderVoll = False
                    for v in i.werte:
                        
                        istKandidat = True
                        for z in Felder:
                            if (i.box == z.box and i != z) and type(z.werte) != int:  #freie Felder in gleicher Box
                                if v in z.werte:
                                    istKandidat = False
                        if istKandidat == True:
                            i.werte = v
                            print("Feld",Felder.index(i),"ausgefüllt mit Schritt 2-2, nochmal Schritt 1!")
                            Schritt1(Felder)              #NEUER EINTRAG, NOCHMAL NEU RAUSSTREICHEN -> SCHRITT 1
                            break
                        
                    
                        istErneutKandidat = True
                        for z in Felder:
                            if (i.zeile == z.zeile) and i != z and type(z.werte) != int: #freie Felder in gleicher Box
                                if v in z.werte:
                                    istErneutKandidat = False
                        if istErneutKandidat == True:
                            i.werte = v
                            print("Feld",Felder.index(i),"ausgefüllt mit Schritt 2-3, nochmal Schritt 1!")
                            Schritt1(Felder)              #NEUER EINTRAG, NOCHMAL NEU RAUSSTREICHEN -> SCHRITT 1
                            break
                        
                        
                        istNochmalKandidat = True
                        for z in Felder:
                            if (i.spalte == z.spalte and i != z) and type(z.werte) != int: #freie Felder in gleicher Box
                                if v in z.werte:
                                    istNochmalKandidat = False
                        if istNochmalKandidat == True:
                            i.werte = v
                            print("Feld",Felder.index(i),"ausgefüllt mit Schritt 2-4, nochmal Schritt 1!")
                            Schritt1(Felder)              #NEUER EINTRAG, NOCHMAL NEU RAUSSTREICHEN -> SCHRITT 1
                            break
        
        if FelderVoll == True:
            print("2 Stufe fertig")
            break
                                    
                  
        
    #----------------Ergebnisaufbereitung------------------------
    
    print(FelderVoll,n)
    Solutionmatrix = [list(range(1,10)),    #Blankomatrix
                      list(range(1,10)),
                      list(range(1,10)),
                      list(range(1,10)),
                      list(range(1,10)),
                      list(range(1,10)),
                      list(range(1,10)),
                      list(range(1,10)),
                      list(range(1,10))] 
    for l in Felder:
        Solutionmatrix[l.zeile][l.spalte] = l.werte    #Füllen Blankomatrix mit echten Werten
    
    SolutionAsNumpy = np.array(Solutionmatrix)    #umformatierung für schöneres print
    
    #-------------Lösungskontrolle-------------------------
    
    def Solutioncheck(felder):
        allesOkay = True
        for i in range(1,10):
            kontrollliste = list(range(1,10))
            for k in felder:
                if k.box == i:
                    if k.werte in kontrollliste:
                        kontrollliste.remove(k.werte)
                    else:
                        allesOkay = False
            if len(kontrollliste) == 0:
                break
            else:
                allesOkay = False
        
        for i in range(0,9):
            kontrollliste = list(range(1,10))
            for k in felder:
                if k.zeile == i:
                    if k.werte in kontrollliste:
                        kontrollliste.remove(k.werte)
                    else:
                        allesOkay = False
            if len(kontrollliste) == 0:
                break
            else:
                allesOkay = False
        
        for i in range(0,9):
            kontrollliste = list(range(1,10))
            for k in felder:
                if k.spalte == i:
                    if k.werte in kontrollliste:
                        kontrollliste.remove(k.werte)
                    else:
                        allesOkay = False
            if len(kontrollliste) == 0:
                break
            else:
                allesOkay = False
                
        if allesOkay == True:
            print("CHECK OKAY")
        else:
            print("ERROR IN SOLUTION")
                
        
    #------------   DONE -----------------    
    
    print(SolutionAsNumpy)  
    Solutioncheck(Felder)
    
          
```
